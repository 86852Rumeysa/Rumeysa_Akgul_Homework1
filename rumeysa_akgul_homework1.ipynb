```python
import thinkplot
import thinkstats2
```

```python
marriage=resp.fmarno
hist = thinkstats2.Hist(marriage, label='Number of Marriage Done by Respondents')
thinkplot.Hist(hist)
thinkplot.Config(xlabel='Number of Marriage', ylabel='Number of Respondents')
```

```python
never_married = resp[resp.fmarno == 0]
married = resp[resp.fmarno >=1]

nevermarried = thinkstats2.Hist(never_married.totincr, label='Never Married')
Married = thinkstats2.Hist(married.totincr, label='Married')
```

```python
width = 0.45
thinkplot.PrePlot(2)
thinkplot.Hist(nevermarried, align='right', width=width)
thinkplot.Hist(Married, align='left', width=width)
thinkplot.Config(xlabel='Total Income', ylabel='Number of People')
```
